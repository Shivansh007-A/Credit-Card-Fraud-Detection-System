```python
df["Class"].value_counts() # 0 for normal transaction and 1 for fraud transaction
```

```python
fraud_data = df.loc[(df["Class"]==1)]
normal_data = df.loc[(df["Class"]==0)]
```

```python
df.groupby("Class").mean()
```

```python
normal_sample = normal_data.sample(n=500) #creating a new dataset that contains some random samples from the noraml_data dataset
```

```python
new_df = pd.concat([normal_sample,fraud_data]) # created new_df dataset that contains fraud and normal data in a balanced way
new_df
```

```python
X = new_df.drop("Class",axis=1) #stored all new_df dataset in X except Class column's values
y = new_df["Class"] #stored Class column's values in y 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
var = LogisticRegression()
```

```python
var.fit(X_train,y_train) #training our data
```

```python
print(len(var.predict(X_test)))
```

```python
var.score(X_test,y_test) # Accuracy score of testing data
```

```python
var.score(X_train,y_train) #Accuracy score of training data
```
